```python
page_limit = 50
lgas = ['wyndham', 'nillumbik', 'monash', 'bayside_vic', 'knox', 'yarra_city', 'kingston', 'darebin',
 'yarra_ranges', 'banyule', 'moreland_city', 'manningham', 'whitehorse', 'brimbank', 'boroondara_city', 'mornington_peninsula',
 'frankston_city', 'glen_eira', 'city_of_whittlesea', 'latrobe', 'stonnington', 'casey', 'cardinia', 'city_of_port_phillip', 'melbourne_city', 'maroondah_city'] 

combined_df_array = []

for lga in lgas:
    
    lga_df_array = []
    page = 1

    while page <= page_limit:
        page_df = query_api(lga,page)
        page_df.rename(columns={'authority.full_name': 'lga_fullname'}, inplace=True)
        page_df['lga_urlname'] = lga

        #<< may need to add a wait
        print(f'Retrieved page {page} from {lga}')

        lga_df_array.append(page_df)

        page = page + 1
        
    combined_df_array.append(pd.concat(lga_df_array))

combined_df = pd.concat(combined_df_array).drop_duplicates()
```
